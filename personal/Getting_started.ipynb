```julia
# Original node positions (x,y)
x = [0, 0, 0, 0]
y = [0, 1, 2, 3]

# Fake displacements (to test plotting)
dx = [0.0, 0.1, 0.15, 0.2]
dy = [0.0, -0.5, -0.1, -0.8]

# Plot before and after
plot(x, y, marker=:circle, label="Original", legend=:bottomright)
plot!(x .+ dx, y .+ dy, marker=:star, label="Displaced")
```

```julia
#define funct
linear(x, m, b) = m*x + b
#test
println(linear(1, 2, 3))
```

```julia
#define quadratic
quadratic(x, a, b, c) = a * x^2 + b*x + c
#test
println(quadratic(2, 1, 1, 1))
```

```julia
#define linspace
xspan = -1.0:0.5:1.0
plot(xspan, [linear.(xspan, 2, 1), quadratic.(xspan, 1, 2, 1)], 
    label=["linear", "quadratic"], xlabel="x", ylabel="f(x)")
```
